```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv('/content/drive/My Drive/관광/최종2.csv')
```

# 전처리

```python
quartile_1 = df['D_TRA1_S_Day'].quantile(0.25)
quartile_3 = df['D_TRA1_S_Day'].quantile(0.75)
IQR = quartile_3 - quartile_1
df_result = df[(df['D_TRA1_S_Day'] < (quartile_1 - 1.5 * IQR)) | (df['D_TRA1_S_Day'] > (quartile_3 + 1.5 * IQR))]
df.drop(df_result.index, axis=0,inplace=True)
```


+ 최종본 복사를 위한 df2


```python
df2=df.copy()
```

# 최초변수선택

```python
df.columns
```

```python
df.drop(columns=['SAGE','D_TRA1_1_SPOT', 'D_TRA1_1_Code', 'D_TRA1_1_Q6_Ar', 'D_TRA1_1_Q6','A10A_4','C_TOTAL','DQ2A','DQ3A_1','DQ3A_2','DQ3A_3','DQ3A_4'],inplace=True)
```

+ ['D_TRA1_CASE', 'BARA', 'BJOB','DQ3','A1_1','A4_1'] 
+ 명목형 변수들 Onehot Encoding

```python
OneHot = ['D_TRA1_CASE', 'BARA', 'BJOB','DQ3','A1_1',
          'A4_1']


df3 = pd.get_dummies(df, drop_first=True,columns=OneHot)
```

```python
변수리스트=df3.columns.tolist()
```

```python
df3.columns
```

+ Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df3[변수리스트] = scaler.fit_transform(df3[변수리스트])
```

```python
feature = df3[변수리스트]
```

# Clustering

+ 최적 군집값 도출 n=7

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
import seaborn as sns
distortions = []
for i in range(1, 20):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(df3[변수리스트])
    distortions.append(km.inertia_)
plt.plot(range(1, 20), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
model = KMeans(n_clusters=7,algorithm='auto')

model.fit(feature)

predict = pd.DataFrame(model.predict(feature))

predict.columns=['predict']
```

```python
df2.reset_index(inplace=True,drop =True)
df3.reset_index(inplace=True,drop=True)
```

```python
df3['cluster'] = predict
df2['cluster'] = predict
df2.to_csv('/content/drive/My Drive/관광/cluster_result.csv')
```
